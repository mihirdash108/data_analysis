```python
i=0
j=0
x=len(apu_contact)
y=len(ad_contact)
apu_fi=[]
apu_fid=[]
apu_fin=[]
mm=""
for i in range(x):
    for j in range(y):
        if(apu_contact[i]==ad_contact[j]):
            if(ad_channel[j]=="workshop"):
                apu_fi.append(ad_channel[j])
                apu_fid.append(ad_order_at[j])
                apu_fin.append(ad_course_name[j])
                break
            elif(ad_channel[j]=="MEMBERSHIP" or ad_channel[j]=="LIVE_COURSE" or ad_channel[j]=="MEMBERSHIP_BUNDLE"):
                apu_fi.append(mm)
                apu_fid.append(mm)
                apu_fin.append(mm)
                break
```
